```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.dineout.co.in/bangalore-restaurants')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
first_title = soup.find('a',class_="restnt-name ellipsis")
first_title
```

```python
first_title.text
```

```python
loc = soup.find('div',class_='restnt-loc ellipsis')
loc.text
```

```python
sta = soup.find('span',class_="double-line-ellipsis")
sta.text
```

```python
sta = soup.find('span',class_="double-line-ellipsis")
sta.text.split('|')[0]
```

```python
titles = []
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    titles.append(i.text)
    
    
    
    
titles    
```

```python
titles
```

```python
location = []
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
    
    
    
    
location   
```

```python
price = []

for i in soup.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text.split('|')[0])
    
    
    
    
price   
```

```python
images = []
for i in soup.find_all("img", class_="no-img"):
    images.append(i.get('data-src'))
images    
```

```python
print(len(titles),len(location),len(price),len(images))
```

```python
import pandas as pd
df = pd.DataFrame({'Titles':titles,'Location':location,'Price':price,'Images_url':images})
df
```
